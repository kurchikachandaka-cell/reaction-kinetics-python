```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Time (s)
time = np.array([0, 20, 40, 60, 80])

# Concentration (mol/L)
concentration = np.array([0.200, 0.163, 0.134, 0.110, 0.090])

# Take natural log
ln_conc = np.log(concentration)

# Plot ln[A] vs time
plt.scatter(time, ln_conc)
plt.xlabel("Time (s)")
plt.ylabel("ln[A]")
plt.title("First-Order Kinetics Plot")
plt.show()

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(time, ln_conc)

k = -slope
print("Rate constant k =", k, "s^-1")
print("R-squared =", r_value**2)
```
